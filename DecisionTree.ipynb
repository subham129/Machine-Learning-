```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('salaries.csv')
```

```python
x = df.drop('salary_more_then_100k', axis = 1)
y = df['salary_more_then_100k']
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
x['company_n'] = le.fit_transform(df['company'])
x['job_n'] = le.fit_transform(df['job'])
x['degree_n'] = le.fit_transform(df['degree'])
```

```python
x
```

```python
x.drop(columns= ['company', 'job', 'degree'], inplace = True)
```

```python
y = le.fit_transform(y)
```

```python
x
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier
cif = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
cif.fit(x_train, y_train)
cif.score(x_test,y_test)
```
