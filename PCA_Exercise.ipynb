```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df['ChestPainType'].unique()
```

```python
df['RestingECG'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse=False)
le = LabelEncoder()
```

```python
df_encoded = pd.DataFrame(ohe.fit_transform(df[['ChestPainType']]),columns=ohe.get_feature_names_out(['ChestPainType']))
```

```python
df_encoded
```

```python
pd.concat([df,df_encoded],axis=1)
```

```python
df.columns
```

```python
df_encoded2 = pd.DataFrame(ohe.fit_transform(df[['ST_Slope']]),columns=ohe.get_feature_names_out(['ST_Slope']))
```

```python
df_encoded2
```

```python
df['RestingECG'] = le.fit_transform(df['RestingECG'])
```

```python
df.head()
```

```python
df = pd.concat([df,df_encoded,df_encoded2],axis=1)
```

```python
df
```

```python
df.drop(columns=['ChestPainType','ST_Slope'],inplace=True)
```

```python
df['Sex'] = le.fit_transform(df['Sex'])
```

```python
df['ExerciseAngina'] = le.fit_transform(df['ExerciseAngina'])
```

```python
df.head()
```

```python
x = df.drop(columns=['HeartDisease'])
y = df['HeartDisease']
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = cross_val_score(LogisticRegression(max_iter=1000),x,y,cv=10).mean()
rf = cross_val_score(RandomForestClassifier(),x,y,cv=10).mean()
```

```python
print(lr)
print(rf)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
x_pca = PCA(0.95).fit_transform(x)
```

```python
x_pca.shape
```

```python
cross_val_score(LogisticRegression(max_iter=1000),x_pca,y,cv=10).mean()
```
