```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df['number'].unique()
```

```python
df['number_numeric'] = pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
df['number_categorical'] = np.where(df['number_numeric'].isnull(),df['number'],np.nan)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')
df['cain_cat'] = df['Cabin'].str[0]
```

```python
df.head()
```

```python
df['Ticket'].unique()
```

```python
df['ticket_num'] = df['Ticket'].apply(lambda x: x.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
