```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked',])
```

```python
df.head()
```

```python
dum = pd.get_dummies(df['Sex'])
df = pd.concat([df,dum],axis=1)
```

```python
df.head()
```

```python
df.drop(columns=['Sex'],inplace=True)
```

```python
x = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
x['Age'] = x.Age.fillna(x.Age.mean())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x.isna().sum()
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(model,x,y,cv=10,scoring='accuracy').mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
cross_val_score(RandomForestClassifier(),x,y,cv=100,scoring='accuracy').mean()
```
