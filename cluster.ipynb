```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df['sepal length (cm)'],df['sepal width (cm)'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df['sepal length (cm)'] = scaler.fit_transform(df[['sepal length (cm)']])
df['sepal width (cm)'] = scaler.fit_transform(df[['sepal width (cm)']])
```

```python
df.head()
```

```python
kn = KMeans(n_clusters=3)
pred = kn.fit_predict(df[['sepal length (cm)','sepal width (cm)']])
```

```python
df['cluster'] = pred
```

```python
df.head()
```

```python
df1 = df[df['cluster']==0]
df2 = df[df['cluster']==1]
df3 = df[df['cluster']==2]
```

```python
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='green')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='red')
plt.scatter(df3['sepal length (cm)'],df3['sepal width (cm)'],color='black')
plt.scatter(kn.cluster_centers_[:,0],kn.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```
